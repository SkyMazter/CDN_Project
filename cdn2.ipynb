```python
from ipaddress import ip_address, IPv4Address, IPv6Address, IPv4Network, IPv6Network
from fabrictestbed_extensions.fablib.fablib import FablibManager as fablib_manager

fablib = fablib_manager()
                     
fablib.show_config();
```

```python
slice_name = 'cdn2'
[site1, site2] = fablib.get_random_sites(count=2)
print(f"Sites: {site1}, {site2}")

server_name = 'server'
client_name = 'client'
network_name1 = 'net1'
network_name2 = 'net2'
server_nic_name = 'server_nic'
client_nic_name = 'client_nic'
router_nic_name1 = 'router_nic_1'
router_nic_name2 = 'router_nic_2'


image = 'default_ubuntu_20'
```

```python
try:
    # Create Slice
    slice = fablib.new_slice(name=slice_name)

    # Networks
    net1 = slice.add_l2network(name=network_name1, subnet=IPv4Network("192.168.1.0/24"))
    net2 = slice.add_l2network(name=network_name2, subnet=IPv4Network("192.168.2.0/24"))

    # Server Node
    server = slice.add_node(name=server_name, site=site1, image=image)
    server_iface = server.add_component(model='NIC_Basic', name=f"{server_nic_name}_1").get_interfaces()[0]
    server_iface.set_mode('auto')
    net1.add_interface(server_iface)
    
    # Client Node
    client = slice.add_node(name=client_name, site=site2, image=image)
    client_iface = client.add_component(model='NIC_Basic', name=f"{client_nic_name}_1").get_interfaces()[0]
    client_iface.set_mode('auto')
    net2.add_interface(client_iface)

    # Router Node
    router = slice.add_node(name="router", site=site1, image=image)
    router_iface1 = router.add_component(model='NIC_Basic', name=router_nic_name1).get_interfaces()[0]
    router_iface2 = router.add_component(model='NIC_Basic', name=router_nic_name2).get_interfaces()[0]
    router_iface1.set_mode('auto')
    router_iface2.set_mode('auto')
    net1.add_interface(router_iface1)
    net2.add_interface(router_iface2)

    # Submit Slice Request
    slice.submit()
except Exception as e:
    print(f"Exception: {e}")
    
```
